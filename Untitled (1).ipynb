```python
import os
import pandas as pd
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import seaborn as sns
import cv2
from tensorflow.keras.models import load_model
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
# 1. Define the directory path (Fixed with 'r' for Windows path)
data_dir = r'C:\Users\Hp\Downloads\archive\dataset2-master\dataset2-master\images\TEST'

# 2. Define the class labels
class_labels = ['eosinophil', 'lymphocyte', 'monocyte', 'neutrophil']

# 3. Initialize lists to hold file paths and labels
filepaths = []
labels = []

# 4. Loop through each class directory and gather file paths and labels
# Added a check to make sure the directory exists to avoid errors
if os.path.exists(data_dir):
    for label in class_labels:
        class_dir = os.path.join(data_dir, label)
        
        if os.path.exists(class_dir):
            for file in os.listdir(class_dir):
                if file.endswith('.jpeg') or file.endswith('.png') or file.endswith('.jpg'):
                    filepaths.append(os.path.join(class_dir, file))
                    labels.append(label)
        else:
            print(f"Warning: Folder not found for label: {label}")
else:
    print("Error: The main data_dir does not exist. Please check the path.")

# 5. Create a DataFrame from the file paths and labels
bloodCell_df = pd.DataFrame({
    'filepaths': filepaths,
    'labels': labels
})

# 6. Shuffle the DataFrame
if not bloodCell_df.empty:
    bloodCell_df = bloodCell_df.sample(frac=1).reset_index(drop=True)
    print(f"Total images found: {len(bloodCell_df)}")
    print(bloodCell_df.head())
else:
    print("No images found! Check if your dataset folders are inside 'dataset2-master'.")
```

```python
import matplotlib.pyplot as plt
import numpy as np

def show_knee_images(image_gen):
    test_dict = image_gen.class_indices
    classes = list(test_dict.keys())
    images, labels = next(image_gen)
    plt.figure(figsize=(20, 20))
    length = len(labels)
    
    if length < 25:
        r = length
    else:
        r = 25
        
    for i in range(r):
        plt.subplot(5, 5, i + 1)
        image = (images[i] + 1) / 2
        plt.imshow(image)
        index = np.argmax(labels[i])
        class_name = classes[index]
        plt.title(class_name, color="green", fontsize=16)
        plt.axis('off')
        
    plt.show()

# Isko run karne ke liye (ensure 'train' generator is defined):
# show_knee_images(train)
```

```python
# 1. Sabse pehle data ko split karo (Agar pehle nahi kiya hai)
train_df, val_df = train_test_split(bloodCell_df, train_size=0.8, shuffle=True, random_state=42)

# 2. Train Generator define karo
# MobileNetV2 use kar rahe ho toh preprocess_input zaroori hai
train_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)

# 3. 'train' variable define karo (Jo show_knee_images function mang raha hai)
train = train_datagen.flow_from_dataframe(
    dataframe=train_df,
    x_col='filepaths',
    y_col='labels',
    target_size=(224, 224), # MobileNetV2 ka default size
    color_mode='rgb',
    class_mode='categorical',
    batch_size=32,
    shuffle=True
)

print("Train generator is ready!")
```

```python
show_knee_images(train)
```

```python
# Data ko split karna
train_images, test_images = train_test_split(bloodCell_df, test_size=0.3, random_state=42)
train_set, val_set = train_test_split(bloodCell_df, test_size=0.2, random_state=42)

# Shapes print karna results check karne ke liye
print(train_set.shape)
print(test_images.shape)
print(val_set.shape)
print(train_images.shape)
```

```python
# Image Data Generator initialize karna
image_gen = ImageDataGenerator(preprocessing_function=tf.keras.applications.mobilenet_v2.preprocess_input)

# Train set ke liye generator
train = image_gen.flow_from_dataframe(
    dataframe=train_set,
    x_col="filepaths",
    y_col="labels",
    target_size=(244, 244),
    color_mode='rgb',
    class_mode="categorical",
    batch_size=8,
    shuffle=False
)

# Test set ke liye generator
test = image_gen.flow_from_dataframe(
    dataframe=test_images,
    x_col="filepaths",
    y_col="labels",
    target_size=(244, 244),
    color_mode='rgb',
    class_mode="categorical",
    batch_size=8,
    shuffle=False
)

# Validation set ke liye generator
val = image_gen.flow_from_dataframe(
    dataframe=val_set,
    x_col="filepaths",
    y_col="labels",
    target_size=(244, 244),
    color_mode='rgb',
    class_mode="categorical",
    batch_size=8,
    shuffle=False
)
```

```python
import tensorflow as tf
from tensorflow import keras

model = keras.models.Sequential([
    # Input Layer and First Convolutional Block
    keras.layers.Conv2D(filters=128, kernel_size=(8, 8), strides=(3, 3), activation='relu', input_shape=(224, 224, 3)),
    keras.layers.BatchNormalization(),

    # Second Convolutional Block
    keras.layers.Conv2D(filters=256, kernel_size=(5, 5), strides=(1, 1), activation='relu', padding="same"),
    keras.layers.BatchNormalization(),
    keras.layers.MaxPool2D(pool_size=(3, 3)),

    # Third Convolutional Block (with 1x1 convolutions)
    keras.layers.Conv2D(filters=256, kernel_size=(3, 3), strides=(1, 1), activation='relu', padding="same"),
    keras.layers.BatchNormalization(),
    keras.layers.Conv2D(filters=256, kernel_size=(1, 1), strides=(1, 1), activation='relu', padding="same"),
    keras.layers.BatchNormalization(),
    keras.layers.Conv2D(filters=256, kernel_size=(1, 1), strides=(1, 1), activation='relu', padding="same"),
    keras.layers.BatchNormalization(),

    # Fourth Convolutional Block
    keras.layers.Conv2D(filters=512, kernel_size=(3, 3), activation='relu', padding="same"),
    keras.layers.BatchNormalization(),
    keras.layers.MaxPool2D(pool_size=(2, 2)),

    # Fifth Convolutional Block
    keras.layers.Conv2D(filters=512, kernel_size=(3, 3), activation='relu', padding="same"),
    keras.layers.BatchNormalization(),

    # Sixth Convolutional Block
    keras.layers.Conv2D(filters=512, kernel_size=(3, 3), activation='relu', padding="same"),
    keras.layers.BatchNormalization(),
    keras.layers.MaxPool2D(pool_size=(2, 2)),

    # Seventh Convolutional Block
    keras.layers.Conv2D(filters=512, kernel_size=(3, 3), activation='relu', padding="same"),
    keras.layers.BatchNormalization(),
    keras.layers.MaxPool2D(pool_size=(2, 2)),

    # Fully Connected Layers
    keras.layers.Flatten(),
    keras.layers.Dense(1024, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1024, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(4, activation='softmax')
])

# Compile the model
model.compile(
    loss='categorical_crossentropy',
    optimizer=tf.optimizers.SGD(learning_rate=0.001),
    metrics=['accuracy']
)

# Display the model architecture
model.summary()
```

```python
history = model.fit(train,epochs = 5,validation_data = val,verbose =1)
```
